```python
import pandas as pd
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
faq_data = {
    "Question": [
        "What is Java?",
        "Who developed Java?",
        "What is the JVM?",
        "What is the JDK?",
        "What is the JRE?",
        "What is Object-Oriented Programming?",
        "What is a class in Java?",
        "What is an object in Java?",
        "What are the features of Java?",
        "What is inheritance in Java?"
    ],

    "Answer": [
        "Java is a high-level, object-oriented programming language used to build desktop, web, and mobile applications.",
        "Java was developed by James Gosling at Sun Microsystems in 1995.",
        "JVM stands for Java Virtual Machine. It allows Java programs to run on different operating systems.",
        "JDK stands for Java Development Kit. It contains tools needed to develop and compile Java applications.",
        "JRE stands for Java Runtime Environment. It provides the libraries and JVM required to run Java programs.",
        "Object-Oriented Programming (OOP) is a programming paradigm based on classes and objects.",
        "A class is a blueprint for creating objects in Java.",
        "An object is an instance of a class that contains data and methods.",
        "Java features include platform independence, object-oriented programming, security, robustness, multithreading, and portability.",
        "Inheritance is an OOP concept where one class acquires the properties and methods of another class."
    ]
}

faq = pd.DataFrame(faq_data)

faq
```

```python
nltk.download('punkt_tab')
```

```python
import string

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

faq['Processed_Question'] = faq['Question'].apply(preprocess_text)

faq[['Question', 'Processed_Question']]
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(faq['Processed_Question'])

print(tfidf_matrix.shape)
```

```python
def chatbot(user_question):
    processed_question = preprocess_text(user_question)

    user_vector = vectorizer.transform([processed_question])

    similarity = cosine_similarity(user_vector, tfidf_matrix)

    index = similarity.argmax()

    return faq.iloc[index]['Answer']
```

```python
user_question = input("Ask a question: ")

response = chatbot(user_question)

print("Chatbot:", response)
```

```python
while True:
    user_question = input("You: ")

    if user_question.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    response = chatbot(user_question)

    print("Chatbot:", response)
```

```python
import gradio as gr
def chatbot_interface(user_question):
    return chatbot(user_question)

interface = gr.Interface(
    fn=chatbot_interface,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Ask a question..."
    ),
    outputs="text",
    title="FAQ Chatbot",
    description="Ask any question related to the FAQ dataset."
)

interface.launch()
```
